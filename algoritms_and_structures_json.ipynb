```python
import json
from collections import Counter
from re import findall
d = json.load(open('tolstoy.json', 'r')) # файл предварительно был скопирован на компьютер
result = {}
words = []

for st in d['INTRODUCTION']['INTRODUCTION']['INTRODUCTION']: # подсчет слов для Introduction
    words+= findall(r'\w+',st.lower())
result['INTRODUCTION'] = Counter(words).copy()
    

for book in d.keys(): # подсчет слов для каждой части, помеченной римской цифрой
    result[book]={}
    if not('INTRODUCTION'== book):
        for part in d[book].keys():
            result[book][part]={}
            for num in d[book][part].keys():
                words=[]
                for st in d[book][part][num]:
                    words += findall(r'\w+',st.lower())
                result[book][part][num] = Counter(words)

for book in result.keys(): # подсчет слов для каждой части. Будет сохранено с ключом 'All Part'
    if not('INTRODUCTION'== book):
        for part in result[book].keys():
            all_part_dict = Counter({})             # подсчет слов для каждой части
            for num in result[book][part].keys():
                for word in result[book][part][num].keys():
                    all_part_dict[word] += result[book][part][num][word] # используется свойство класса Counter
                    # при отсутствии ключа в словаре возвращается 0
            result[book][part]['All Part'] = all_part_dict.copy()
            
for book in result.keys(): #  подсчет слов каждой книги. Будет сохранено с ключом 'All Book'
    if not('INTRODUCTION'== book):
        all_book_dict = Counter({})  #  подсчет слов каждой книги
        for part in result[book].keys():
            for word in result[book][part]['All Part'].keys():
                all_book_dict[word]+= result[book][part]['All Part'][word] # используется свойство класса Counter
                # при отсутствии ключа в словаре Counter возвращает 0
        result[book]['All Book'] = all_book_dict.copy()

whole_count = Counter({})
for book in result.keys(): # подсчет слов всей книги. Будет сохранено с ключом 'All'
    if not('INTRODUCTION'== book):
        for word in result[book]['All Book'].keys():
            whole_count[word] += result[book]['All Book'][word]
    else:
        for word in result[book].keys():
            whole_count[word] += result[book][word]
result['All'] = whole_count.copy()
print('Cтруктура полученного файла:')
print()
for book in result.keys():
    print(book)
    if not(book=='All') and not(book=='INTRODUCTION'):
        for part in result[book].keys():
            print('\t'+part)
            if not part == 'All Book':
                for num in result[book][part].keys():
                    print('\t\t'+num)
    print()
with open('answer.json', 'w') as f:
    json.dump(result, f)
```
