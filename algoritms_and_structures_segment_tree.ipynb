```python
import numpy as np

test = [2,10,-16,5,1,7]
def full_of_tree(test): # создание дерева отрезков на основе данного массива 
    main_minimum = -10**(6) # число, заведомо меньшее всех элементов массива
    size = len(test)
    k = 0
    while(size > 2**k):
        k+=1
    size = (2**(k+1) - 1) # size - размер дерева отрезков,  k+1 - количество слоев
    seg_tree = [main_minimum]*size
    num = 0
    for i in range(len(test)): # заполнение последнего слоя
        seg_tree[len(seg_tree)-2**k+i] = test[i]
    # заполнение более высоких слоев
    last_slice_0 = len(seg_tree)
    last_slice_e = len(seg_tree)
    for j in range(k):
        last_slice_e = last_slice_0 - 1
        last_slice_0 = last_slice_0 - 2**(k-j)
        for i in range(last_slice_0,last_slice_e,2):
            seg_tree[(i-1)//2] = max(seg_tree[i],seg_tree[i + 1])
    
    # вывод массива и дерева
    test = test + [main_minimum]*(2**k - len(test))
    print('test')
    print(test) # test array
    print('segment tree')
    print(seg_tree) # segment tree
    print()
    return seg_tree,k

seg_tree,k = full_of_tree(test) # создание дерева

def max_of_two_val(right,left): # для max_of_segment. Возвращает индекс максимального элемента или -1
    if left == -1 and right == -1:
        return -1
    if left == -1 and not(right==-1):
        return right
    if right== -1 and not(left == -1):
        return left
    if seg_tree[left] > seg_tree[right]:
        return left
    else: 
        return right

def interval_to_index(a,b): # для max_of_segment. Конвертирует обозначение сегмента в индекс массива seg_tree
    s = k - int(np.log2(b-a+1))
    if s == 0:
        index = 0
    else:
        index_of_slice = 2**s - 1 # индекс начала слоя в массиве
        size_of_one_segment = 2**(k-s)
        pos = a//size_of_one_segment # позиция в массиве элемента внутри слоя
        index = index_of_slice + pos
    return index # индекс  нужного сегмента в массиве

def max_of_segment(a,b,mmin=0,mmax=2**k-1): # возвращает индекс максимального элемента в сегменте (a,b)
    if (a<=mmin) and (mmax<=b): 
        return interval_to_index(mmin,mmax)
    if (b < mmin) or (mmax < a): 
        return -1 
    if (mmin<a) and (mmax<=b):
        left = max_of_segment(a,b,mmin,mmin+(mmax-mmin)//2) # индекс максимума левого поддерева
        right = max_of_segment(a,b,mmax-(mmax-mmin)//2,mmax) # индекс максимума правого поддерева
        return max_of_two_val(right,left) # возвращает индекс максимального элемента
    if (a<=mmin) and (b<mmax):
        left = max_of_segment(a,b,mmin,mmin+(mmax-mmin)//2)
        right = max_of_segment(a,b,mmax-(mmax-mmin)//2,mmax)
        return max_of_two_val(right,left)
    if (mmin<a) and (b<mmax):
        left = max_of_segment(a,b,mmin,mmin+(mmax-mmin)//2)
        right = max_of_segment(a,b,mmax-(mmax-mmin)//2,mmax)
        return max_of_two_val(right,left)

print('SEARCH OF MAXIMUM')
print()
for a,b in [(0,7),(0,3),(0,4),(1,2),(0,1),(1,1)]: # поиск максимума для разных сегментов
    print('Search of maximum of segment ({},{})'.format(a,b))
    print('array:',end=' ')
    print(test)
    print('segment:',end=' ')
    print(test[a:b+1])
    if not(max_of_segment(a,b) == -1):
        print('maximum of segment:',end=' ')
        print(seg_tree[max_of_segment(a,b)])
    else:
        print('Диапазон выходит за границы сегментов')
        print('границы сегментов ({},{})'.format(0,2**k-1))
    print()
def add_number(seg_tree,a,b,num,k): # добавляет к сегменту (a,b) целое число num. k+1 - количество слоев дерева
    # не используется lazy алгоритм, при каждом добавлении полный пересчет всех элементов дерева.
    test = seg_tree[-2**k:len(seg_tree)].copy()
    print('before operation')
    print('test',end=' ')
    print(test)
    print('segment tree',end=' ')
    print(seg_tree)
    print('after {} was added to segmtent ({},{}): {}'.format(num,a,b,test[a:b+1]),end='->')
    for i in range(a,b+1):
        test[i]+=num
    print('{}'.format(test[a:b+1]))
    seg_tree,k = full_of_tree(test)
    return seg_tree
print('ADD A NUMBER')
print()
for a,b in [(0,7),(0,3),(0,4),(1,2),(0,1),(1,1)]: # добавление 4 к разным сегментам
    seg_tree = add_number(seg_tree,a,b,4,k)

        
```
